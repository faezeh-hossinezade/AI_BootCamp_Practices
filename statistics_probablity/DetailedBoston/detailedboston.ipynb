<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در این بخش از شما می‌خواهیم با استفاده از روش 
<code>iqr</code>
داده‌های پرت را <b style="color:#0099cc">جایگزین</b> نمایید. بدین منظور در هر ستون بجز ستونی که تنها مقادیر ۰ و ۱ را می‌پذیرد رویه زیر را دنبال کنید. 
نخست مقدار چارک اول و سوم را به دست آورید. سپس مقدار <code>iqr</code> را محاسبه کرده و مقادیر ۱.۵ برابر <code>iqr</code> بالاتر از چارک سوم و پایین‌تر از چارک اول را به ترتیب حد بالا و پایین در نظر بگیرید. در نهایت هر داده‌ای خارج از محدوده‌ی مجاز را با حد متناظر (داده‌های بزرگ با حد بالا و داده‌های کوچک با حد پایین) جایگزین کنید.
</font>
</p>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
<b style="color:#0099cc">نکته:</b>
تغییرات خود را روی یک کپی از دیتافریم اصلی (<code>df</code>) با نام <code>df_rep</code> اعمال کنید.
</font>
</p>


```python
import pandas as pd
import numpy as np


df_rep = df.copy()

numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
# This line of code creates a list of column names containing numerical data.
# Breakdown:
# 1. `df.select_dtypes(include=['number'])`: Selects all columns from the DataFrame `df` that have a numerical data type (e.g., int, float).
# 2. `.columns`: Extracts the column names from the selected DataFrame.
# 3. `.tolist()`: Converts the Series of column names into a Python list for easier manipulation.
# In essence, the `numeric_cols` list will hold the names of all numerical columns in the DataFrame `df`. This list can be used for various operations on numerical columns, such as statistical calculations, plotting, and more.
numeric_cols = [col for col in numeric_cols if len(df[col].unique()) > 2]


for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    

    df_rep[col] = np.where((df[col] < lower_bound), lower_bound, np.where((df[col] > upper_bound), upper_bound, df[col]))
    # Replace outliers in column 'col' with their respective bounds

binary_cols = [col for col in df.columns if col not in numeric_cols]
df_rep[binary_cols] = df[binary_cols]
```


<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">

مدیریت داده‌های پرت
۲

</h2>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در این قسمت می‌خواهیم دو روش <b style="color:#0099cc">حذف</b> داده‌های پرت به کمک <code>iqr</code> و <code>z score</code> را با هم مقایسه کنیم.
برای هر کدام از نمونه‌ها مقدار 
<code>z score</code>
را برای ستون 
<code>dis</code>
حساب کنید و داده‌هایی که مقدار مطلق 
<code>z score </code>
آن‌ها بیشتر از ۳ است را حذف کنید.
سپس واریانس داده‌های به‌دست‌آمده را با با حالتی که با روش 
<code>iqr</code>
داده‌های پرت را حذف کنیم با هم مقایسه کنید (یعنی نیاز است با هر دو حالت به‌صورت جداگانه داده‌های پرت را حذف کرده و در نهایت نتیجه را مقایسه کنید).
</font>
</p>

$$‌ z = \frac{(x-\mu)}{\sigma} $$
$$ \small{z: \text{z-score}} $$
$$ \small{x: \text{value}} $$
$$ \small{\mu: \text{mean}} $$
$$ \small{\sigma: \text{standard deviation}} $$

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
<b style="color:#0099cc">نکته:</b>
تغییرات خود برای حذف داده‌های پرت به کمک روش <code>z score</code> را روی یک کپی از دیتافریم اصلی (<code>df</code>) با نام <code>df_zscore</code>، و برای حذف داده‌های پرت به کمک <code>iqr</code> روی <code>df_iqr</code> اعمال کنید.
</font>
</p>


```python
mean = df['dis'].mean()
std = df['dis'].std()
df['z_score'] = (df['dis'] - mean) / std
df_zscore = df.copy()

df_zscore = df_zscore[abs(df_zscore['z_score']) <= 3]

df_zscore = df_zscore.drop('z_score', axis=1)

df_zscore
```

```python
Q1 = df['dis'].quantile(0.25)
Q3 = df['dis'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_iqr = df.copy()
df_iqr = df_iqr[(df_iqr['dis'] >= lower_bound) & (df_iqr['dis'] <= upper_bound)]
df_iqr
```

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
لازم است اختلاف واریانس دو مجموعه داده ی جدید که با دو روش مختلف فیلتر شده اند را بدست آورید و آن را در متغیر var_difference ذخیره کنید.
</font>
</p>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
این متغیر توسط تصحیح خودکار بررسی خواهد شد و درست بودن آن صرفا به معنای درست بودن اختلاف واریانس هاست و دلیل آن برای چک کردن درست بودن دیتاست های بخش قبلیست و کسب نمره ی آن بخش کمی از نمره ی نهایی سوال است.
</font>
</p>

```python
var_zsore= df_zscore ['dis'].var()
var_iqr=df_iqr['dis'].var()

var_zsore,var_iqr
```

```python
var_difference=np.abs(var_zsore-var_iqr)
var_difference
```

```python
import scipy.stats as stats
```

```python
u_statistic_ad, p_value_ad = stats.mannwhitneyu(older_houses, younger_houses, alternative='less')
p_value_ad
```

```python
alpha=0.05

if p_value_ad<alpha:
    print('reject the null hypothesis')
else:
    print('fail to reject the null hypothesis')
```
