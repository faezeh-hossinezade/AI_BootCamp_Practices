       - ستون  <code>Country_Region</code>: مشخص است که نام کشور یا منطقه را مشخص می‌کند. دقت کنید که شاید لازم باشد بررسی کنید که آیا تعداد سطرها با تعداد کشورها برابر است و به بیان دیگر به ازای هر منطقه یا کشور فقط یک سطر دیتا وجود دارد یا خیر.
        <br>
        - ستون‌های <code>Long</code> و <code>Lat</code>: طول و عرض جفرافیایی کشور/منطقه را بیان می‌کند.

        - ستون <code>Confirmed</code>: تعداد موارد تاییدشده‌ی ابتلا به ویروس

        - ستون <code>Recovered</code>: تعداد موارد بهبودیافته

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
قسمت اول‌
</font>
</h2>

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
     در ابتدا می‌خواهیم ببینیم ۱۰ کشور با بیشترین تعداد موارد  کدام‌اند. برای این کار می‌توانیم از
ستون  <code>Incident_Rate</code> استفاده کنیم. البته برای این‌که به‌نوعی مجموع تقریبی موارد را از این ستون محاسبه کنیم می‌توانیم از متد <code>apply</code>  استفاده کنیم و ستونی جدید به نام <code>Total_Cases</code> اضافه کنیم که از حاصل ضرب ۱۰۰۰۰۰ در مقادیر ستون <code>Incident_Rate</code> توسط تابع <code>apply</code> به‌دست می‌آیند.
<br>
پاسخ شما باید دارای ۱۰ ردیف و ۲ ستون باشد که نام کشور/منطقه و <code>Total_Cases</code> را شامل می‌شود و آن را در فایلی به نام <code>top10cases.csv</code> ذخیره کنید. 
</font>
</p>

```python
covid_data['Total_Cases'] = covid_data['Incident_Rate'].apply(lambda x: round(x*100000,0))
dta= covid_data.groupby('Country_Region').sum().reset_index()
top10cases=dta[['Country_Region','Total_Cases']].sort_values('Total_Cases',ascending=False)
top10cases=top10cases.head(10)
top10cases.to_csv('top10cases.csv', header=False)
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
قسمت دوم
</font>
</h2>

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
در ادامه می‌خواهیم ببینیم چه کشورهایی بیشترین نرخ ریکاوری را داشته‌اند. بدین منظور قصد داریم شاخصی تعریف کنیم تحت عنوان <code>Recovery_Rate</code> که از حاصل تقسیم تعداد بهبودیافتگان به تعداد موارد تایید شده بدست می‌آید. البته در ابتدا قصد داریم کشورها/مناطقی که موارد تایید شده، فوت شده و بهبودیافته‌ی آن‌ها صفر گزارش شده است را درنظر نگیریم.
پس ردیف‌هایی که مقدار هر سه ستون <code>Confirmed</code>، <code>Deaths</code> و <code>Recovered</code> آن‌ها صفر است را از دیتافریم اصلی حذف کنید. سپس می‌خواهیم در بین کشورهایی که موارد تاییدشده به ابتلای آنها از میانه‌ی این شاخص بیشتر است ببینیم کدام‌شان نرخ ریکاوری بهتری دارد. بدین منظور در بین این کشورها، فهرست آنهایی که نرخ ریکاوری آنها از میانگین این نرخ برای همه کشورها بیشتر است را به صورت نزولی  
 در فایلی به نام <code>recovery_rate.csv</code> ذخیره کنید. پس فیلتر اولیه بر اساس میانه‌ی تعداد موارد تایید شده‌ی مبتلایان و فیلتر دوم بر اساس میانگین نرخ ریکاوری است.
<br>
توجه داشته باشید که فایل شما باید شامل دو ستون باشد که به ترتیب نام کشور/منطقه و نرخ ریکاوری باشد و ترتیب آن به صورت نزولی بر اساس نرخ ریکاوری.
</font>
</p>

```python
dta = covid_data.groupby('Country_Region')[['Recovered','Confirmed','Deaths']].sum().reset_index()
dta = dta.drop(dta[(dta['Recovered']==0)&(dta['Confirmed']==0)&(dta['Deaths']==0)].index)
dta.dropna(how="any", inplace=True)
conf_median=dta['Confirmed'].median()
dta['Recovery_Rate']=dta['Recovered']/dta['Confirmed']
average_recovery_rate=dta['Recovery_Rate'].mean()
result = dta[(dta['Recovery_Rate']>average_recovery_rate) & (dta['Confirmed']>conf_median)][['Country_Region', 'Recovery_Rate']].sort_values('Recovery_Rate', ascending=False)
result.to_csv('recovery_rate.csv', header=False)
result
```

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
اکنون می‌خواهیم اوضاع سایر کشورها را با کشور خودمان مقایسه‌ای کنیم! از این روی می‌خواهیم ببینیم کدام کشورها با وجود این‌که مبتلایان بیشتری از ما داشته‌اند، مرگ و میر کمتری ثبت کرده‌اند.
<br>
پاسخ را در فایلی به نام <code> vsiran.csv </code> ذخیره کنید. فایل پاسخ باید حاوی ۴ ستون باشد که نام کشور/منطقه، موارد تایید شده، موارد فوت‌شده و موارد بهبودیافته را به‌صورت نزولی بر اساس تعداد موارد تایید شده در بر داشته باشد.
</font>
</p>

```python
dta=covid_data.groupby('Country_Region').sum().reset_index()
vs_death_iran=int(dta['Deaths'][dta['Country_Region']=='Iran'])
vs_conf_iran=int(dta['Confirmed'][dta['Country_Region']=='Iran'])
result=dta[(dta['Deaths']<vs_death_iran)&(dta['Confirmed']>vs_conf_iran)]
vsiran=result[['Country_Region','Confirmed','Deaths','Recovered']].sort_values('Confirmed',ascending=False)
vsiran.to_csv('vsiran.csv', header= False)
vsiran
```

در ادامه می‌خواهیم ادعای سخنگوی وزارت بهداشت کشوری را ارزیابی کنیم که ادعا می‌کند کشورش با توجه به آمار فوتی‌هایشان در این روز، جزو ۲۰ درصد برتر جهان است (هنوز در آن زمان عبارتی مثل «کشورهای برتر یک درصد جهان» کشف نشده بود :) ).
<br>
ما تعداد مرگ‌های گزارش‌شده را به تنهایی کافی نمی‌دانیم و فرض کنید برای راستی‌آزمایی این ادعا می‌خواهیم از شاخص ساده‌ای چون <b style="color:#0099cc">نسبت تعداد متوفیان به مبتلایان</b> استفاده کنیم. البته برای اینکه اینقدر هم دیگر ساده نباشد (!)، می‌خواهیم کشورهایی که مبتلایان تایید شده‌ی کمی در این روز داشته‌اند را در نظر نگیریم و برای این کار ۲۰ درصد کشورهایی که کمترین مبتلایان را داشته‌اند را نیز کنار می‌گذاریم و سپس شاخص مربوطه را برای کشورهای باقی‌مانده لحاظ می‌کنیم.
<br>
مقدار شاخص را در ستونی به نام <code>death_ratio</code> ذخیره کرده و
فهرست ۲۰ درصد برتر را به همراه مقدار شاخص ذکر شده در فایلی به نام 
    <code>20percent.csv</code> ذخیره کنید.

</font>
</p>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<b style="color:orange">توجه:</b>
آن دسته از ردیف‌هایی که مقدار ستون <code>Confirmed</code> آن‌ها برابر صفر است را در نظر نگرفته و حذف کنید.
</font>
</p>

<p dir=rtl style="direction: rtl;text-align: justify;line-height:200%;font-family:vazir;font-size:medium">
<b style="color:green">راهنمایی:</b>
دیتافریم حاصل شامل ۳۰ ردیف خواهد بود.
</font>
</p>

```python
df = covid_data.groupby(['Country_Region']).sum().reset_index()
df.drop(df[df['Confirmed']==0].index,inplace=True)
df.dropna()
threshold = df['Confirmed'].quantile(0.2)
df = df[df['Confirmed'] > threshold]
df['death_ratio'] = df['Deaths'] / df['Confirmed']
df = df[['Country_Region','death_ratio']].sort_values('death_ratio', ascending=True)
top_20 = df.head(int(0.2 * len(df)))
top_20.to_csv("20percent.csv", header=False)
top_20
```

```python
import zlib
import zipfile

def compress(file_names):
    print("File Paths:")
    print(file_names)
    compression = zipfile.ZIP_DEFLATED
    with zipfile.ZipFile("result.zip", mode="w") as zf:
        for file_name in file_names:
            zf.write('./' + file_name, file_name, compress_type=compression)

file_names = ["top10cases.csv", "recovery_rate.csv", "vsiran.csv", "20percent.csv", "covid19.ipynb"]
compress(file_names)
```
