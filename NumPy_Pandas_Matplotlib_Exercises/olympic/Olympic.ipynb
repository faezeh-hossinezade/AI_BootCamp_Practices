```python
gold_medals = data[data['Medal'] == 'Gold']
gold_medal_counts = gold_medals['Country'].value_counts().reset_index(name='Gold_Count')
top_10_countries = gold_medal_counts.head(10)
top_10_countries
```

```python
fig_top10, ax_top10 = plt.subplots(figsize=(10, 6))
fig_top10.patch.set_facecolor('#EAF4f4')
ax_top10.set_facecolor('#EAF4f4')

x = top_10_countries['Country']
y = top_10_countries['Gold_Count']
bars = ax_top10.bar(x, y, color='#005f73')

ax_top10.spines['top'].set_visible(False)
ax_top10.spines['right'].set_visible(False)
ax_top10.spines['left'].set_visible(False)

ax_top10.tick_params(left=False, length=0)
ax_top10.set_xticks(range(len(x)))
ax_top10.tick_params(axis='x', rotation=45, labelrotation=45, labelright=True, labelsize=10, labelcolor='#22333b')

ax_top10.set_yticklabels(ax_top10.get_yticks(), fontsize=10, color='#22333b')
ax_top10.set_ylabel('Number of Gold Medals', fontsize=10, color='#22333b')
ax_top10.set_title('Top 10 Countries with the Most Gold Medals', fontsize=12, color='#22333b')

ax_top10.set_ylim(top=max(y) + 100)
ax_top10.grid(False, axis='x')
```

```python
men_medals = data[data['Gender'] == 'Men']
women_medals = data[data['Gender'] == 'Women']
men_medal_counts = men_medals['Year'].value_counts().reset_index(name='men_Count')
sorted_men_maedal_counts = men_medal_counts.sort_values('Year', ascending=False)
women_medal_counts = women_medals['Year'].value_counts().reset_index(name='women_Count')
sorted_women_maedal_counts = women_medal_counts.sort_values('Year', ascending=False)
sorted_women_maedal_counts
```

```python
sorted_men_maedal_counts
```

```python
grouped_data = data.groupby('Year')['Gender'].value_counts().unstack()
# print(grouped_data)
grouped_data = grouped_data.fillna(0)
# print(grouped_data)

fig_gen, ax_gen = plt.subplots(figsize=(10, 6))
fig_gen.patch.set_facecolor('#EAF4f4')
ax_gen.set_facecolor('#EAF4f4')

men_color = '#005f73'
women_color = '#94d2bd'

ax_gen.plot(grouped_data.index, grouped_data['Men'], color=men_color, marker='o', label='Men')
ax_gen.plot(grouped_data.index, grouped_data['Women'], color=women_color, marker='o', label='Women')

ax_gen.fill_between(grouped_data.index, grouped_data['Women'], color=women_color, alpha=0.6)
ax_gen.fill_between(grouped_data.index, grouped_data['Women'], grouped_data['Men'], color=men_color, alpha=0.6)

ax_gen.spines['top'].set_visible(False)
ax_gen.spines['right'].set_visible(False)
ax_gen.spines['left'].set_visible(False)
ax_gen.tick_params(axis='both', direction='out', length=0)
ax_gen.set_xlim(1896, 2012)
ax_gen.yaxis.tick_right()

plt.title('Male and Female Won Medals over the Years', fontsize=12, color='#22333b')
plt.ylabel('Number of Won Medals', fontsize=10, color='#22333b')

ax_gen.yaxis.set_tick_params(color='#22333b', labelcolor='#22333b')
ax_gen.xaxis.set_tick_params(color='#22333b', labelcolor='#22333b')

legend = plt.legend(facecolor='#EAF4f4', loc='upper right')
legend.get_frame().set_alpha(None)

plt.show()
```

```python
filtered_data = data[data['Country'] == 'SVK']
grouped_data = filtered_data.groupby('Sport')['Medal'].value_counts().unstack()
# print(grouped_data)
grouped_data = grouped_data.fillna(0)
# print(grouped_data)
grouped_data
```

```python

import numpy as np
import matplotlib.pyplot as plt

fig_SVK, ax_SVK = plt.subplots(figsize=(10, 8))
ax_SVK.set_facecolor('#EAF4f4')
fig_SVK.patch.set_facecolor('#EAF4f4')

wedge_colors = ['#d8f3dc', '#b7d4c7', '#95d5b2', '#74c69d', '#52b788', '#40916c']
medal_colors = {'Gold': '#E9C46A', 'Silver': '#E0E1DD', 'Bronze': '#CD7F32'}
#  This defines a dictionary mapping medal names (Gold, Silver, Bronze) to their corresponding colors.
outer_radius = 1
# Radius of the outer circle.
inner_radius = 0.7
# Radius of the inner circle.
width = 0.35
edgecolor = 'w'
# Color of the edges between slices (set to white here).

# Outer donut plot
outer_medals = grouped_data.sum(axis=1)
outer_labels = grouped_data.index
wedges, outer_texts = ax_SVK.pie(outer_medals, radius=outer_radius, labels=outer_labels,
                                 colors=wedge_colors, wedgeprops=dict(width=width, edgecolor=edgecolor))

# Inner donut plot
inner_data = grouped_data.values.flatten()
inner_labels = grouped_data.columns.repeat(grouped_data.shape[0])
inner_medals = inner_data[inner_data > 0]
inner_labels = inner_labels[inner_data > 0]
inner_colors = [medal_colors.get(medal) for medal in inner_labels]
inner_wedges, inner_texts = ax_SVK.pie(inner_medals, radius=inner_radius, labels=inner_labels,
                                       colors=inner_colors, labeldistance=0.75 if inner_radius == 0.7 else None,
                                       wedgeprops=dict(width=width, edgecolor=edgecolor))
plt.setp(inner_wedges, width=width * 0.7)

# Plot settings
plt.title('Count of Medals in each Sport', fontsize=12, color='#22333b')



for text in inner_texts:
    text.set_horizontalalignment('center')

# Show the plot

ax_SVK.set_facecolor('#EAF4F4')
fig_SVK.patch.set_facecolor('#EAF4F4')
plt.show()
```
